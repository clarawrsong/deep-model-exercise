```python
def sample_data_sensor(df, types, p_num, s_types, trial): # sensor, 'exc01', 1, 'normal', ''
    data = df[ (df['p_num'] == p_num) 
              & (df['s_types'] == s_types ) 
              & (df['types'] == types ) 
              & (df['trial'] == trial)] # trial은 slow and fast구분 안함. 
    #print(len(data))
    data = data.reset_index()
    return data

def sample_data_video(df, types, v_types,  num): # video, 'exc01', 'YouTube', 1
    data = df[ (df['v_types'] == v_types ) 
                & (df['types_x'] == types ) 
                & (df['trial'] == num)]
    #print(len(data))
    data = data.reset_index()
    return data
```

```python
train_X = []
train_Y = []
for i in range(0, len(train_video)):
    exc_name = train_video.types[i]
    video_types = train_video.v_types[i]
    exc_num = train_video.types_n[i]
    #print(i, exc_name)
    for j in range(1, train_video.trial[i]+1):
        trial = j
        tmp = sample_data_video(video, exc_name, video_types, trial) # video, 'exc01', 'YouTube', 1
        x  = list(savgol_filter(signal.resample(tmp.w_x_right_r_unit_v_new, 60), 19, 1))
        y  = list(savgol_filter(signal.resample(tmp.w_y_right_r_unit_v_new, 60), 19, 1))
        z  = list(savgol_filter(signal.resample(tmp.w_z_right_r_unit_v_new, 60), 19, 1))
        x1 = list(savgol_filter(signal.resample(tmp.e_x_right_r_unit_v_new, 60), 19, 1))
        y1 = list(savgol_filter(signal.resample(tmp.e_y_right_r_unit_v_new, 60), 19, 1))
        z1 = list(savgol_filter(signal.resample(tmp.e_z_right_r_unit_v_new, 60), 19, 1))
        tmp_list = [x, y, z, x1, y1, z1]
        merged = list(itertools.chain(*tmp_list))
        merged = sklearn.preprocessing.minmax_scale(merged, feature_range=(0, 1), axis=0, copy=True)
        train_X.append(merged)
        train_Y.append(exc_num)
        print(i, exc_name, trial)

    
```

```python
test_X = []
test_Y = []
for i in range(0, len(test_sensor_ground)):
    p_num = test_sensor_ground.p_num[i]
    exc_name = test_sensor_ground.types[i]
    sensor_types = test_sensor_ground.s_types[i]
    exc_num = test_sensor_ground.types_n[i]
    #print(i, exc_name)
    for j in range(1, test_sensor_ground.trial[i]+1):
        trial = j
        try:
            tmp = sample_data_sensor(sensor, exc_name, p_num, sensor_types, trial)
            x  = list(savgol_filter(signal.resample(tmp.w_x_right, 60), 19, 1))
            y  = list(savgol_filter(signal.resample(tmp.w_y_right, 60), 19, 1))
            z  = list(savgol_filter(signal.resample(tmp.w_z_right, 60), 19, 1))
            x1 = list(savgol_filter(signal.resample(tmp.e_x_right, 60), 19, 1))
            y1 = list(savgol_filter(signal.resample(tmp.e_y_right, 60), 19, 1))
            z1 = list(savgol_filter(signal.resample(tmp.e_z_right, 60), 19, 1))
            tmp_list = [x, y, z, x1, y1, z1]
            merged = list(itertools.chain(*tmp_list))
            merged = sklearn.preprocessing.minmax_scale(merged, feature_range=(0, 1), axis=0, copy=True)
            test_X.append(merged)
            test_Y.append(exc_num)
            print('p_num: ',p_num, i, exc_name, trial)
        except:
            print("ERROR!!!", 'p_num: ',p_num, i, exc_name, trial)
    
```

```python
test_X_1 = []
test_Y_1 = []
for i in range(0, len(test_sensor_real)):
    p_num = test_sensor_real.p_num[i]
    exc_name = test_sensor_real.types[i]
    sensor_types = test_sensor_real.s_types[i]
    exc_num = test_sensor_real.types_n[i]
    #print(i, exc_name)
    for j in range(1, test_sensor_real.trial[i]+1):
        trial = j
        try:
            tmp = sample_data_sensor(sensor, exc_name, p_num, sensor_types, trial) # video, 'exc01', 'YouTube', 1
            x  = list(savgol_filter(signal.resample(tmp.w_x_right, 60), 19, 1))
            y  = list(savgol_filter(signal.resample(tmp.w_y_right, 60), 19, 1))
            z  = list(savgol_filter(signal.resample(tmp.w_z_right, 60), 19, 1))
            x1 = list(savgol_filter(signal.resample(tmp.e_x_right, 60), 19, 1))
            y1 = list(savgol_filter(signal.resample(tmp.e_y_right, 60), 19, 1))
            z1 = list(savgol_filter(signal.resample(tmp.e_z_right, 60), 19, 1))
            tmp_list = [x, y, z, x1, y1, z1]
            merged = list(itertools.chain(*tmp_list))
            merged = sklearn.preprocessing.minmax_scale(merged, feature_range=(0, 1), axis=0, copy=True)
            test_X_1.append(merged)
            test_Y_1.append(exc_num)
            print('p_num: ',p_num, i, exc_name, trial)
        except:
            print("ERROR!!!", 'p_num: ',p_num, i, exc_name, trial)

    
```
